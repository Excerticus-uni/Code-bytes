Limits

```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.size']=14
# Generate x values from 0 to
# 100000000000 (10 zeros) and 1000 steps
x=np.linspace(0,1e20,10000)
y_num=x+1
y_den=x+1e6
y=y_num/y_den
plt.plot(x,y_num,'b-') #Plot lines
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')
# Display the legend. Labels between the $
#signs are in latex. The "r" prefix indicates
#Python to expect latex language
#for the legend text
plt.legend([r'$x+1$',r'$x+10^6$',\
r'$\frac{x+1}{x+10^6}$'])
# Display the axis labels
plt.xlabel('x')
plt.ylabel('y')
#For very large values, numerator and
#denominator are identical!
#The head-start 10^6 in the denominator
#has no consequences in the limi
```

Exponential 

```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
x=np.linspace(0,5,1000)
y=np.exp(x)
plt.plot(x,y,'b-')
plt.plot(x,x**2,'g--')
plt.plot(x,x,'r--')
plt.legend([r'$y=e^x$',r'$y=x^2$',\
r'$y=x$'])
plt.xlabel('x');plt.ylabel('y')
```

```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
x=np.linspace(0,100,1000)
y=np.exp(x)
plt.plot(x,y,'b-')
plt.plot(x,x**2,'g--')
plt.plot(x,x**10,'r--')
plt.legend([r'$y=e^x$',r'$y=x^2$',\
r'$y=x^{10}$'])
plt.xlabel('x');plt.ylabel('y')
```
